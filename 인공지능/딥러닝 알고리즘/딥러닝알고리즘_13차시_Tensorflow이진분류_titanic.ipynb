```python
import pandas as pd
```

```python
raw_titanic = pd.read_csv('train.csv', index_col='PassengerId')
raw_titanic.head()
```

```python
raw_titanic.tail()
```

```python
raw_titanic.drop(['Name', 'Ticket', 'Fare', 'Cabin', 'Embarked'], axis=1, inplace=True)
raw_titanic.head()
```

```python
titanic_data = pd.get_dummies(raw_titanic, columns=['Sex'], dtype=int)
titanic_data.head(10)
```

```python
titanic_data['age_label']=pd.cut(titanic_data.Age, bins=[0, 2, 17, 65, 99], labels=['Baby', 'Child', 'Adult', 'Elderly'])
titanic_data.head(10)
```

```python
titanic_data.drop(['Age'], axis=1, inplace=True)
titanic_data.head(10)
```

```python
titanic_data.fillna('Adult', inplace=True)
titanic_data.head(10)
```

```python
titanic_data = pd.get_dummies(titanic_data, columns=['age_label'], dtype=int)
titanic_data.head(10)
```

```python
X = titanic_data.iloc[:, 1:]
y = titanic_data.iloc[:, 0]
```

```python
X.head()
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
from tensorflow.keras import models, layers, optimizers
```

```python
model = models.Sequential()
model.add(layers.Dense(24, input_dim=9, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(12, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
# 직접 계산할 줄도 알아야하는 매우 중요한 부분!!
model.summary()
```

## << 매우 중요 >>  
dense_1의 param 개수 : 9(입력) * 24(출력) + 24   
(편향=여기서는 출력의 개수와 같음)  
(가중치는 9 * 24)  
dense_2의 param 개수 : 24(입력) * 12(출력) + 12(편향)  
dense_3의 param 개수 : 12(입력) * 1(출력) + 1(편향)  


```python
model.compile(loss='binary_crossentropy', optimizer=optimizers.Adam(learning_rate=0.005), metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=50, batch_size=10)

score = model.evaluate(X_test, y_test)
print('Test Accuracy:', score[1])
```
