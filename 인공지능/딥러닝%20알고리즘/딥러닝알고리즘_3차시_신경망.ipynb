3. 신경망  
* 인공 뉴런  
 - 뉴런(node)으로의 입력(Feature) : X = (x_1, x_2, ..., x_n)
 - 각 입력에 곱해지는 가중치(Weight) : W = (w_1, w_2, w_3, ..., w_n)
 - 결과 u = sum(from k=1 to n) (x_k * w_k) + b
 - 위의 결과 u를 활성화함수 f에 입력   
 => 출력 y = f(u) = f(sum(from k=1 to n) (x_k * w_k) + b)  
 => u = XW + b, y = f(XW + b)  

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X1 = np.arange(-1.0, 1.0, 0.2)
X2 = np.arange(-1.0, 1.0, 0.2)
```

```python
Z = np.zeros((10, 10))
```

```python
W = np.array([2.5, 3.0])
bias = np.array([0.1])
```

```python
for i in range(10):
  for j in range(10):
    X = np.array([X1[i], X2[j]])
    u = np.dot(X, W.T) + bias
    y = 1/(1 + np.exp(-u))
    Z[j][i] = y
```

```python
plt.imshow(Z, 'gray', vmin=0.0, vmax=1.0)
plt.colorbar()
plt.show()
```

```python
# 가중치 값의 부호 바꾸기
W = np.array([-2.5, -3.0])
for i in range(10):
  for j in range(10):
    X = np.array([X1[i], X2[j]])
    u = np.dot(X, W.T) + bias
    y = 1/(1 + np.exp(-u))
    Z[j][i] = y
plt.imshow(Z, 'gray', vmin=0.0, vmax=1.0)
plt.colorbar()
plt.show()
```
