```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv('Electric_Production.csv',index_col=0)
data.head()
```

```python
data.index = pd.to_datetime(data.index)
data.index
```

```python
data.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
#위의 그래프의 이동평균을 구하시오.(기간은 30일로 한다.)
```

```python
value = data[['IPG2211A2N']]
value.rolling(30).mean().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=30);
```

```python
#위의 그래프의 차분그래프를 그려보시오.
```

```python
data.diff().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
#위의 그래프의 acf , pacf의 곡선을 그려보시오
```

```python
#ACF , PACF곡선 그리기
plot_acf(data)
plot_pacf(data)
plt.show()
```

```python
#위의 그래프에서 차분을 한번 하고 나서 acf와 pacf를 그려보시오
```

```python
#차분곡선 그래프
diff_1=data.diff(periods=1).iloc[1:]
diff_1.plot()
plot_acf(diff_1)
plot_pacf(diff_1)
plt.show()
```
