```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv('Electric_Production.csv',index_col=0)
data.head()
```

```python
data.index = pd.to_datetime(data.index)
data.index
```

```python
data.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
#   .( 30 .)
```

```python
value = data[['IPG2211A2N']]
value.rolling(30).mean().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=30);
```

```python
#   .
```

```python
data.diff().plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
#  acf , pacf  
```

```python
#ACF , PACF 
plot_acf(data)
plot_pacf(data)
plt.show()
```

```python
#      acf pacf 
```

```python
# 
diff_1=data.diff(periods=1).iloc[1:]
diff_1.plot()
plot_acf(diff_1)
plot_pacf(diff_1)
plt.show()
```
