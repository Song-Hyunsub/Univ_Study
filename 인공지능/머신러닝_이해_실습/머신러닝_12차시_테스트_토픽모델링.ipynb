```python
from sklearn.datasets import fetch_20newsgroups

dataset = fetch_20newsgroups(shuffle=True, random_state=1, remove=('headers', 'footers', 'quotes'))

documents = dataset.data

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
no_features = 1000

# LDA can only use raw term counts for LDA because it is a probabilistic graphical model

tf_vectorizer = CountVectorizer(max_df=0.95, min_df=2, max_features=no_features, stop_words='english')

tf = tf_vectorizer.fit_transform(documents)

tf_feature_names = tf_vectorizer.get_feature_names()
```

```python
tf_feature_names
```

```python
# 위에서 얻은 1000개의 단어를 가지고 뉴스그룹의 토픽모델링 해보기
#LDA방식을 사용하고 learning_method는 online방식으로 epoch는 5로 한다.
#토픽의 개수는 20개로 정한다.
```

```python
from sklearn.decomposition import NMF, LatentDirichletAllocation

no_topics = 20

# Run LDA

lda = LatentDirichletAllocation(n_topics=no_topics, max_iter=5, learning_method='online').fit(tf)
```

```python
lda
```

```python
#20개의 토픽에 대하여 상위10개의 단어를 출력해보기
```

```python
no_top_words = 10

for topic_idx, topic in enumerate(lda.components_):
    print('topic:' , topic_idx)
    print(" ".join([tf_feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))
```
