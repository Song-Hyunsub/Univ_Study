```python
import numpy as np

X = np.random.randint(5, size=(6, 100))
y = np.array([1, 2, 3, 4, 5, 6])
```

```python
X
```

```python
y
```

```python
#나이브 베이즈 모델을 만들고 위의 데이터를 훈련데이터로 삼아서 훈련
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB() # 나이브 베이즈 모델
clf.fit(X, y)
```

```python
#4번째 데이터의 class를 예측
```

```python
print(clf.predict(X[3].reshape(1,100)))
```
