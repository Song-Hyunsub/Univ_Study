```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('Dataset_spine.csv')
```

```python
df
```

```python
df = df.drop(['Unnamed: 13'], axis=1)
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df.drop(['Col7','Col8','Col9','Col10','Col11','Col12'], axis=1)
```

```python
df.head()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
y = df['Class_att']
x = df.drop(['Class_att'], axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.25, random_state=27)
```

```python
x_train
```

```python
y_train
```

```python
#위의 데이터를 이용하여 신경망 모델을 만들어 보자.
#hidden layer의 크기는 각각 100 , 50 , 80으로 하고 epoch의 수는 500으로 한다.
```

```python
clf = MLPClassifier(hidden_layer_sizes=(100,50,80), max_iter=500)
```

```python
clf
```

```python
#모델을 생성하였으면 모델을 훈련하고 예측을 하여 본다. 예측은 x_test로 한다.
```

```python
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
y_pred == y_test
```

```python
#모두 몇%가 정답으로 맞추었는가?
```

```python
import numpy as np
np.mean(y_pred == y_test)
```
