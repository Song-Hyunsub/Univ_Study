```python
import pandas as pd

# 데이터 제공위치
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# 컬럼명 세팅
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']

# iris데이터를 읽어오기
irisdata = pd.read_csv(url, names=names)
```

```python
irisdata.head()
```

```python
# 입력값의 분리
X = irisdata.iloc[:, 0:4] # X값 : 4개의 속성

# 타겟값인 class의 분리
y = irisdata.select_dtypes(include=[object]) # Y값 : class
```

```python
y.head()
```

```python
X.head()
```

```python
y.Class.unique()
```

=> class 범주형 데이터 3개  
=> 숫자로 바꿔야 함!  

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder() # LabelEndoer => 숫자로 바꿈

y = y.apply(le.fit_transform)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
# train_test_split : train 데이터와 test 데이터를 나누는 것
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
# train 데이터는 보통 70%~80%로 정한다.
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(10, 10, 10), max_iter=1000)
# layer 3, 학습반복 1000번
mlp.fit(X_train, y_train.values.ravel())
```

=> 값을 직접 정해줄 수도 있음

```python
y_train
```

```python
y_train.values.ravel()
```

```python
X_test
```

```python
predictions = mlp.predict(X_test)
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,predictions))   # 혼돈 행렬의 2가지 변수 : 실제값 , 예측값
print(classification_report(y_test,predictions))
```

=> 대각선 방향으로 숫자가 모여있어야 함
